```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.style.use('seaborn-v0_8-darkgrid')

data = {
    'ResNet3D': {'skip': [1, 2, 3, 4], 'accuracy': [43.91, 40.81, 36.88, 32.06], 'color': '#D2691E', 'marker': 'o'},  # Оранжевый
    'Swin Large': {'skip': [1, 2, 3, 4], 'accuracy': [47.21, 54.84, 54.40, 51.33], 'color': '#4169E1', 'marker': 's'},  # Синий
    'MViTv2 Small': {'skip': [1, 2, 3, 4], 'accuracy': [52.02, 53.18, 59.54, 56.90], 'color': '#228B22', 'marker': '^'}  # Зеленый
}

my_results = {
    'Mean Accuracy': 51.53,
    'F1-Score': 50.37,
    'Precision': 54.63,
    'Recall': 51.53
}

fig, ax1 = plt.subplots(1, 1, figsize=(8, 6))

for model_name, model_data in data.items():
    ax1.plot(model_data['skip'], model_data['accuracy'], 
             marker=model_data['marker'], linewidth=3, markersize=10,
             label=f"{model_name} (64 фрейма)", color=model_data['color'])

ax1.scatter([0], [my_results['Mean Accuracy']], s=150, color='#B22222', 
           marker='*', label='Моя Модель (128 фреймов)', 
           edgecolors='black', linewidth=1, zorder=5)

ax1.set_xlabel('Skip между фреймами', fontsize=12, fontweight='bold')
ax1.set_ylabel('Mean Accuracy (%)', fontsize=12, fontweight='bold')
ax1.set_title('Сравнение моделей по точности', fontsize=14, fontweight='bold')
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)
ax1.set_xlim(-0.5, 4.5)
ax1.set_ylim(25, 65)

plt.tight_layout()
plt.show()
```
